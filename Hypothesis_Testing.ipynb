# Alternate Hypothesis(Ha): There is significant difference in the diameter of cutlet between 2 units
#Null Hypothesis(Ho): There is no significant difference in the diameter of cutlet between 2 units
#Test carried out: Two sample t-test at 5% significance level.


```python
data=pd.read_csv('Cutlets.csv')
data
```

```python
from scipy import stats
stats.ttest_ind(data['Unit A'], data['Unit B'])
```

#Conclusion: P val is equal to 0.47 which is greater than 0.05
#Hence we accept Null Hypothesis: There is no significant difference in the diameter of cutlet between 2 units
