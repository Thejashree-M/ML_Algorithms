#VISUALIZATION

```python
rules=association_rules(frequent_items3, metric='confidence', min_threshold=0.5)
rules
```

```python
top_rules=rules.nlargest(10,'lift')
top_rules
```

```python
sns.barplot(x=top_rules['lift'], y=top_rules['antecedents'])
```

```python
#scatterplot of support and confidence
plt.scatter(rules.support, rules.confidence, color='red', alpha=0.9)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Support vs Confidence')
plt.show()
```

```python
pivot_table=rules.pivot_table(index='antecedents', columns='consequents', values='support')
sns.heatmap(pivot_table, annot=True, cmap='Paired')
```

```python
pivot_table1=rules.pivot_table(index='antecedents', columns='consequents', values='lift')
sns.heatmap(pivot_table1, annot=True, cmap='Paired')
```

```python
pivot_table2=rules.pivot_table(index='antecedents', columns='consequents', values='confidence')
sns.heatmap(pivot_table2, annot=True, cmap='Paired')
```
